```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=100, n_classes=2, n_informative=1, n_redundant=0, n_features=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=15)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def Perceptron(X, y):
    w1=w2=b=1
    lr=0.1
    for i in range(100):
        for j in range(X.shape[0]):
            z = w1*X[j][0] + w2*X[j][1] + b
           
            if z*y[j] < 0:
                w1 = w1 + (lr * y[j] * X[j][0])
                w2 = w2 + (lr * y[j] * X[j][1])
                b = b + (lr * y[j])
    return w1, w2, b
           
```

```python
w1, w2, b = Perceptron(X, y)
```

```python
print("w1", w1, "\n","w2",  w2, "\n", "b", b)
```

```python
m =  -(w1/w2)
c = -(b/w2)

print(m,"\n", c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c   

plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color="red")
plt.scatter(X[:, 0], X[:, 1], c=y)
```
